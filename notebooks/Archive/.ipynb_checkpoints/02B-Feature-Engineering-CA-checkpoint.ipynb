### Feature Engineer
- For ZRI there are no such columns as: 
    - `ZRI_1bedroom` through `ZRI_5BedroomOrMore`
- Notably, the following are equivalent to one another:
    - `ZHVI_CondoCoop`, `ZHVI_SingleFamilyResidence` $\leftrightarrow$ `Zri_SingleFamilyResidenceRental`
    - `ZHVI_AllHomes` $\leftrightarrow$ `ZRI_AllHomes`
- There is no equivalent for the folloiwng:
    - `ZRI_AllHomesPlusMultifamily`
    - `Zri_MultiFamilyResidenceRental`

```python
zhvi_cols = [col for col in ca_zips.columns if 'zhvi' in col.lower()]
zri_cols = [col for col in ca_zips.columns if 'zri' in col.lower()]

pprint(zhvi_cols)
pprint(zri_cols)
```

#### EDA prior to feature engineering

```python
# Look at the distribution of the ratio of SFR Homes to MultiFam Rentals
sns.distplot(test_ts['ZHVI_SingleFamilyResidence'] / test_ts['Zri_MultiFamilyResidenceRental'], bins=10)
```

```python
# Randomly select 300 of the possible 1,311 zipcodes 
# 300 would give us a sample that has 95% confidence w/ 5% margin of error agains the total 1,311 possible zipcodes

random_zips = np.random.choice(all_zips, 300)

sample_zips = ca_zips[ca_zips['ZipCode'].isin(random_zips)]

for i, zipcode in enumerate(random_zips):
    fig = plt.figure(1, figsize=(20,40))
    
    sub_df = sample_zips[sample_zips['ZipCode'] == zipcode]
    ratio = sub_df['ZHVI_SingleFamilyResidence'] / sub_df['Zri_MultiFamilyResidenceRental']

    fig.add_subplot(60,5,i+1)
    sns.distplot(ratio)
    plt.axvline(ratio.mean(), c='red')
    plt.axvline(ratio.median(), c='black')
```

```python
# In some cases this ratio is not feasible because there are no values for Zri_MultiFamilyResidenceRental
nine_4515 = ca_zips[ca_zips['ZipCode'] == random_zips[1]]
display(nine_4515['ZHVI_SingleFamilyResidence'] / nine_4515['Zri_MultiFamilyResidenceRental'])
display(nine_4515.info())
```
